### Data Cleaning and Preparation 

**Data Type Adjustments**: Convert columns to appropriate types (e.g., dates to datetime).


```python
# changing integers to floats for consistency:

kick_starter['Calories'] = kick_starter['Calories'].astype('float', errors='ignore')
kick_starter['Antioxidant Score'] = kick_starter['Antioxidant Score'].astype('float', errors='ignore')

# displaying updated dataframe
kick_starter.head()
```



### Data Cleaning and Preparation 

**Feature Engineering**: Create at least one new feature from existing data.

Example: 
extracting the month from a date.
Calculate the percentage.
Combine first and last name columns into a full name column.


```python
# creatng new columns for ratios of X-to-calories
kick_starter['P_to_C'] = (kick_starter['Protein (g)'] / kick_starter['Calories'])
kick_starter['F_to_C'] = (kick_starter['Fiber (g)'] / kick_starter['Calories'])
kick_starter['VC_to_C'] = (kick_starter['Vitamin C (mg)'] / kick_starter['Calories'])
kick_starter['AS_to_C'] = (kick_starter['Antioxidant Score'] / kick_starter['Calories'])

# displaying updated dataframe
kick_starter.head()
```

```python
# Data Cleaning and Preparation: Data Type Adjustments: converting new columns to percentages.  
kick_starter['P_to_C'] = kick_starter['P_to_C'].apply(lambda x: f"{x * 100:.2f}%")
kick_starter['F_to_C'] = kick_starter['F_to_C'].apply(lambda x: f"{x * 100:.2f}%")
kick_starter['VC_to_C'] = kick_starter['VC_to_C'].apply(lambda x: f"{x * 100:.2f}%")
kick_starter['AS_to_C'] = kick_starter['AS_to_C'].apply(lambda x: f"{x * 100:.2f}%")

# displaying updated dataframe
kick_starter.head()
```

```python
# Selection Options: Choose which columns or data to focus on and drop the rest.

# decided to drop "Antioxidant Score" column
kick_starter.drop('AS_to_C', axis=1, inplace=True)
kick_starter.head()
```

```python
# decided to display nutrient ratios by food using line graph

# splitting dataframe into 2  parts to faciliate visualization:
midpoint = len(kick_starter) // 2  
first_half = kick_starter.iloc[:midpoint] 
second_half = kick_starter.iloc[midpoint:]  

# needed more- splitting dataframe into 4 parts to faciliate visualization:
midpoint_first_half = len(first_half) // 2
first_half_part1 = first_half.iloc[:midpoint_first_half]
first_half_part2 = first_half.iloc[midpoint_first_half:]

midpoint_second_half = len(second_half) // 2
second_half_part1 = second_half.iloc[:midpoint_second_half]
second_half_part2 = second_half.iloc[midpoint_second_half:]

# creating 4 subplots 
fig, axes = plt.subplots(2, 2, figsize=(16, 14), sharey=True)

# plotting foods A-C
first_half_part1.plot(ax=axes[0, 0], kind='line', marker='o', title='Foods: A-C')
axes[0, 0].set_xticks(range(len(first_half_part1.index)))
axes[0, 0].set_xticklabels(first_half_part1.index, rotation=65)
axes[0, 0].set_xlabel('Index of Foods')
axes[0, 0].set_ylabel('Nutrients')

# plotting foods C-K
first_half_part2.plot(ax=axes[0, 1], kind='line', marker='o', title='Foods: C-K')
axes[0, 1].set_xticks(range(len(first_half_part2.index)))
axes[0, 1].set_xticklabels(first_half_part2.index, rotation=65)
axes[0, 1].set_xlabel('Index of Foods')

# plotting foods K-P
second_half_part1.plot(ax=axes[1, 0], kind='line', marker='o', title='Foods: K-P')
axes[1, 0].set_xticks(range(len(second_half_part1.index)))
axes[1, 0].set_xticklabels(second_half_part1.index, rotation=65)
axes[1, 0].set_xlabel('Index of Foods')
axes[1, 0].set_ylabel('Nutrients')

# plotting foods R-Z
second_half_part2.plot(ax=axes[1, 1], kind='line', marker='o', title='Foods: R-Z')
axes[1, 1].set_xticks(range(len(second_half_part2.index)))
axes[1, 1].set_xticklabels(second_half_part2.index, rotation=65)
axes[1, 1].set_xlabel('Index of Foods')

# displaying plot
plt.suptitle('Line Plots: Protein-to-Calorie (P_to_C) & Fiber-to-Calories (F_to_C) ratios by Food')
plt.tight_layout()
plt.show()
```

### Analysis and Insights
**Findings**: Summarize any patterns observed and address the main project question.



While the protein:calorie and fiber:calorie ratio positively correlated for a few food items such as asparagus, black beans, bok choy, broccoli, brussel sprouts, cauliflower, collard greens, kale, lentils, mustard greens, parsley, spinach, and Swiss chard, the line plots revealed that many food items did not have a strong correlation in protein:calorie and fiber:calorie ratios. At times, inverse relationship between the two aforementioned factors were even noted.  

Foods that had *high protein:calorie ratios* without an accompanying high fiber:calorie ratio included tuna, green tea, water crust, Greek yogurt, clams, Cremini mushrooms, General mushrooms, nutritional yeast, salmon, and sardines.  

Foods that had *high fiber:calorie ratios* without an accompanying high protein:calorie ratio included seaweed, sauerkraut, cinnamon, endives, jicama, kohlrabi, raspberries, and radishes.  
  
  
The *highest protein:calorie ratios* were noted in lentils, chickpeas, black beans, kidney beans, edamame, mackerel, salmon, tempeh, and sardines.  

Foods that were *high in protein with the lowest calories* included tuna, clams, and Greek yogurt.  

The *highest fiber:calorie ratios* were noted in avocados, chickpeas, black beans, kidney beans, pomegranate, lentils, and edamame.  
  
  
Foods that were *high in fiber with the lowest calories* included Chia seeds, green beans, dragonfruit, raspberries, blackberries, artichokes, jicama, and kohlrabi in descending order.  

Foods that were *high in vitamin C with the lowest calories* included red bell peppers, papaya, pamelo, brussel sprouts, kale, grapefruit, kohlrabi, broccoli, pineapples, oranges, kiwi, and cantaloupe.  

Foods with the *highest antioxidant scores with the lowest calories* included cinnamon, turmeric, cayenne pepper, and ginger in descending order.  

Cinnamon was noted to be both *high in antioxidant score:calorie ratio*  as well as *fiber:calorie ratio*.

### Conclusion and Recommendations 
**Summarize**: Present the main conclusions drawn from the analysis.

The following foods would be recommended for individuals were hoping to optimize their protein and fiber intake while limiting their calories: asparagus, black beans, bok choy, broccoli, brussel sprouts, cauliflower, collard greens, kale, lentils, mustard greens, parsley, spinach, and Swiss chard.  

For individuals interested in conserving calories while optimizing protein intake, we would recommend: tuna, green tea, water crust, Greek yogurt, clams, Cremini mushrooms, General mushrooms, nutritional yeast, salmon, and sardines.  


Given the disparity in the results, it can be important for individuals to prioritize whether optimizing protein or fiber (or other nutrients) is their primary priority.   

There are options for optimizing both protein and fiber intake at higher calorie ratios.  

The incorporation of cinnamon is helpful for individuals who being to maintain higher antioxidant and fiber intakes without increasing their calorie intake. Optimizing antioxidants without increasing calories can also be achieved with incorporation of: turmeric, cayenne pepper, and ginger.
